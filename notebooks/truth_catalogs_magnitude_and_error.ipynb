```python
%matplotlib inline
from __future__ import division, print_function
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
import sys, os, warnings
home = os.path.expanduser("~")+'/'
sys.path.append(home+'git/Python/user_modules/')
import match_coord
```

```python
def plot_histograms(td, dr):
    if dr=='3.1':
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            gmag = np.array(22.5-2.5*np.log10(td['DECAM_FLUX'][:, 1]))
            gmagerr = np.array(1/np.sqrt(td['DECAM_FLUX_IVAR'][:, 1])/td['DECAM_FLUX'][:, 1])
            rmag = np.array(22.5-2.5*np.log10(td['DECAM_FLUX'][:, 2]))
            rmagerr = np.array(1/np.sqrt(td['DECAM_FLUX_IVAR'][:, 2])/td['DECAM_FLUX'][:, 2])
            zmag = np.array(22.5-2.5*np.log10(td['DECAM_FLUX'][:, 4]))
            zmagerr = np.array(1/np.sqrt(td['DECAM_FLUX_IVAR'][:, 4])/td['DECAM_FLUX'][:, 4])
    elif dr=='4.0':
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            gmag = np.array(22.5-2.5*np.log10(td['FLUX_G']))
            gmagerr = np.array(1/np.sqrt(td['FLUX_IVAR_G'])/td['FLUX_G'])
            rmag = np.array(22.5-2.5*np.log10(td['FLUX_R']))
            rmagerr = np.array(1/np.sqrt(td['FLUX_IVAR_R'])/td['FLUX_R'])
            zmag = np.array(22.5-2.5*np.log10(td['FLUX_Z']))
            zmagerr = np.array(1/np.sqrt(td['FLUX_IVAR_Z'])/td['FLUX_Z'])
    else:
        raise ValueError('dr not correct')
        
    bins = np.linspace(19, 27, 100)
    fig, ax = plt.subplots(1, 3, figsize=(14, 4))
    mask = np.isfinite(gmag)
    ax[0].hist(gmag[mask], bins=bins)
    ax[0].set_xlabel('gmag')
    mask = np.isfinite(rmag)
    ax[1].hist(rmag[mask], bins=bins)
    ax[1].set_xlabel('rmag')
    mask = np.isfinite(zmag)
    ax[2].hist(zmag[mask], bins=bins)
    ax[2].set_xlabel('zmag')
    plt.show()

    bins = np.linspace(0, 1, 100)
    fig, ax = plt.subplots(1, 3, figsize=(14, 4))
    mask = np.isfinite(gmagerr)
    ax[0].hist(gmagerr[mask], bins=bins)
    ax[0].set_xlabel('gmagerr')
    mask = np.isfinite(rmagerr)
    ax[1].hist(rmagerr[mask], bins=bins)
    ax[1].set_xlabel('rmagerr')
    mask = np.isfinite(zmagerr)
    ax[2].hist(zmagerr[mask], bins=bins)
    ax[2].set_xlabel('zmagerr')
    plt.show()
```

```python
plot_histograms(td, '4.0')
```
